```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Assuming your dataframe is named 'df'
# Load your data into a pandas DataFrame
# df = pd.read_csv('your_data.csv')

# Count the number of unique users
num_users = len(control['user'].apply(lambda x: x['id']).unique())

# Count the total number of tweets
num_tweets = len(control)

# Count the number of retweets
num_retweets = control['retweeted_status'].apply(lambda x: x is not None).sum()

# Create a bar plot for visualization
data = {'Number of Users': [num_users], 'Number of Tweets': [num_tweets], 'Number of Retweets': [num_retweets]}
df_stats = pd.DataFrame(data)

# Create a table for displaying the counts
table_data = {
    'Metric': ['Number of Users', 'Number of Tweets', 'Number of Retweets'],
    'Count': [num_users, num_tweets, num_retweets]
}

# Set up the layout using gridspec
fig = plt.figure(figsize=(12, 6))
grid = gridspec.GridSpec(1, 2, width_ratios=[2, 1])

# Bar chart
ax1 = plt.subplot(grid[0])
df_stats.plot(kind='bar', ax=ax1)
ax1.set_title('Number of Users, Tweets, and Retweets')
ax1.set_ylabel('Count')
ax1.set_xticklabels([])
ax1.legend()

# Table
ax2 = plt.subplot(grid[1])
ax2.axis('tight')
ax2.axis('off')
table = ax2.table(cellText=[table_data['Count']], colLabels=table_data['Metric'], cellLoc='center', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(12)
table.scale(1.5, 1.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import ast

# Load your data into a pandas DataFrame
# df = pd.read_csv('your_new_data.csv')

# Convert string representations of lists to actual lists
infoOps['hashtags'] = infoOps['hashtags'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
infoOps['urls'] = infoOps['urls'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])

# Extract hashtags and URLs
hashtags = infoOps['hashtags'].apply(lambda x: x if isinstance(x, list) else [])
urls = infoOps['urls'].apply(lambda x: x if isinstance(x, list) else [])

# Flatten lists of hashtags and URLs
hashtags = [tag for sublist in hashtags for tag in sublist]
urls = [url for sublist in urls for url in sublist]

# Create a distribution plot for hashtags if there are hashtags
if hashtags:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=hashtags, order=pd.Series(hashtags).value_counts().index[:10])
    plt.title('Top 10 Most Used Hashtags')
    plt.xlabel('Count')
    plt.ylabel('Hashtag')
    plt.show()
else:
    print("No hashtags found.")

# Create a distribution plot for URLs if there are URLs
if urls:
    plt.figure(figsize=(10, 6))
    sns.countplot(y=urls, order=pd.Series(urls).value_counts().index[:10])
    plt.title('Top 10 Most Used URLs')
    plt.xlabel('Count')
    plt.ylabel('URL')
    plt.show()
else:
    print("No URLs found.")
```

```python
import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load your control (organic) and infoOps dataframes
# df_control = pd.read_csv('control_tweets.csv')
# df_info_ops = pd.read_csv('info_ops_tweets.csv')

# Assuming your dataframes have 'full_text' or 'tweet_text' columns
text_control = control['full_text']  # or df_control['tweet_text']
text_info_ops = infoOps['tweet_text']

# Perform sentiment analysis using TextBlob
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'positive'
    elif analysis.sentiment.polarity < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply sentiment analysis to both datasets
sentiments_control = text_control.apply(get_sentiment)
sentiments_info_ops = text_info_ops.apply(get_sentiment)

# Count the frequency of each sentiment in both datasets
sentiment_control_counts = sentiments_control.value_counts()
sentiment_info_ops_counts = sentiments_info_ops.value_counts()


```

```python
plt.figure(figsize=(8, 6))
sentiment_control_counts.plot(kind='bar',color='black' ,label='Control Tweets', alpha=0.7)
sentiment_info_ops_counts.plot(kind='bar', color='blue', label='InfoOps Tweets', alpha=0.7)
plt.title('Sentiment Distribution Comparison')
plt.xlabel('Sentiment')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load your InfoOps and organic tweets datasets
# df_info_ops = pd.read_csv('info_ops_tweets.csv')
# df_organic = pd.read_csv('organic_tweets.csv')

# Assuming your dataframes have 'full_text' or 'tweet_text' columns
text_info_ops = infoOps['tweet_text']
text_organic = control['full_text']  # or df_organic['tweet_text']

# Combine all text data into a single string for word cloud generation
all_text_info_ops = ' '.join(text_info_ops)
all_text_organic = ' '.join(text_organic)

# Create a WordCloud object for InfoOps tweets
wordcloud_info_ops = WordCloud(width=800, height=400, background_color='white').generate(all_text_info_ops)

# Create a WordCloud object for organic tweets
wordcloud_organic = WordCloud(width=800, height=400, background_color='white').generate(all_text_organic)

# Plot the WordClouds side by side
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.imshow(wordcloud_info_ops, interpolation='bilinear')
plt.title('InfoOps Tweets')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(wordcloud_organic, interpolation='bilinear')
plt.title('Organic Tweets')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
def word_info(response_control):
    tmp = []
    for i in response_control:
        if i.prediction == 'infoOps':
            tmp.append(1)
        else:    
            tmp.append(0)
    return tmp
```

```python
def word(response_control):
    tmp = []
    for i in response_control:
        if i.prediction == 'control':
            tmp.append(1)
        else:    
            tmp.append(0)
    return tmp
```

```python
predictions = []
for i in combined_control:
    response_control = co.classify(
    inputs=i,
    examples=examples,
    )
    tmp = word(response_control)
    predictions = predictions + tmp
```
